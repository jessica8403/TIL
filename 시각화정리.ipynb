서울특별시 출생률  

```python
import folium  
import json
```

```python
geo_path = '/content/drive/MyDrive/seoulsigungu.geojson'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
df3=pd.read_excel('/content/drive/MyDrive/연간평균.xlsx',index_col='지역')
```

```python
df3 #2010년-2020년 출생률 
```

```python
df3.index
```

```python

map = folium.Map(location=[37.5502, 126.982], zoom_start=11,
                tiles='stamentoner')
map.choropleth(geo_data=geo_str,
              data=df3,
              columns=[df3.index, '평균'],
              fill_color='PuRd',
              key_on = 'properties.SIG_KOR_NM')
map
```

```python
#2010-2020년 서울특별시 출생률 
# matplotlib.rcParams['font.family']='NanumGothicCoding'
# matplotlib.rcParams['axes.unicode_minus']=False
# plt.rcParams["figure.figsize"] = (10,5)

# df1.plot()
# plt.show()

df1_1=df1.plot(grid=True,style=['r--*', 'g-o', 'b:*', 'm-.p'])
df1_1.set_xlabel('연도')
df1_1.set_ylabel('출생률')
df1_1.set_title('서울특별시 출생률')
df1_1.plot(size=(10,7))
plt.show()
```

서울특별시 주택매매가격 증감율과 출산 증감율과의 관계 (주택범위-아파트,단독주택,연립주택,다세대주택,다가구주택)

```python
df9=pd.read_excel('/content/drive/MyDrive/연주택매매증가율.xlsx',index_col='연도')
```

```python
df9
```

```python
df10=pd.read_excel('/content/drive/MyDrive/연간출생증감율수정.xls',index_col='연도')
```

```python
df10
```

```python
df10.rename(columns= {'증감율' : '출생증감율'},
    inplace = True)
```

```python
df11=df9.join(df10)
```

```python
df11
```

```python

df11_1=df11.plot(grid=True,style=['r--*', 'g-o', 'b:*', 'm-.p'])
df11_1.set_xlabel('연도')
df11_1.set_ylabel('증감율')

df11_1.set_title('서울특별시 주택출산증감율')
df11_1.plot(size=(10,7))
plt.show()
```
