```python
import tensorflow as tf
import pandas as pd
```

```python
수산물 = pd.read_csv('/content/drive/MyDrive/수산물_위판정보_전처리완료.csv.csv의 사본')
print(수산물.columns)
수산물.head()
```

```python
수산물.head(30)
```

```python
독립 = 수산물[['총판매액']]
종속 = 수산물[['평균가']]
```

```python
print(독립.shape, 종속.shape)
```

```python
X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
model.fit(독립, 종속, epochs=20, verbose=0)
```

```python
print(model.predict(독립[300:350]))
```
